```python
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt
import functions
from functions import factorial, r_factorial, r_factorial_prime, num_derivative
```

```python
def plot_something(x, y, ax, name, color, linestyle="solid", marker=None):
    ax.plot(x, y, label=name, color=color, linestyle=linestyle, marker=marker)
    ax.set_xlabel("x", fontsize=20)
    ax.set_ylabel("y", fontsize=20)
    ax.set_ylim(0, y_max)
    ax.set_xlim(0, x_max)
    ax.tick_params(axis='x', labelsize=14)
    ax.tick_params(axis='y', labelsize=14)



x_max = 10
y_max = 10000
precision = 1000
    
x = np.linspace(0, x_max, precision + 1)
print("x")
print(x)

x_fact = np.linspace(0, x_max, x_max +1)
y_fact = factorial(x_fact)

y_rfac = r_factorial(x)
print()
print("g(x)")
print(y_rfac)

y_rfac_prime = r_factorial_prime(x)
print()
print("g'(x)")
print(y_rfac_prime)


x_rfac_prime_num, y_rfac_prime_num = num_derivative(x, y_rfac)
print()
print("g'(x) numerical")
print(x_rfac_prime_num)
print(y_rfac_prime_num)


y_quart = functions.quartic(x)
y_quart_prime = functions.quart_prime(x)
x_quart_prime_num, y_quart_prime_num = num_derivative(x, y_quart)


fig, ax = plt.subplots(figsize=[15, 10])

#plot_something(x_lin, y_lin, ax, "lin", "red")
plot_something(x_fact, y_fact, ax, "f(x)", "black", marker="o", linestyle="None")
plot_something(x, y_rfac, ax, "g(x)", "red")
plot_something(x, y_rfac_prime, ax, "g'(x)", "orange")
plot_something(x_rfac_prime_num, y_rfac_prime_num, ax, "g'(x) numerical", "purple")
#plot_something(x, y_quart, ax, "quart", "blue")
#plot_something(x, y_quart_prime, ax, "quart_prime", "green")
#plot_something(x_quart_prime_num, y_quart_prime_num, ax, "quart_prime_num", "grey")




plt.legend(loc="upper left", prop={'size': 30})
#plt.savefig("figures/figure1.png", bbox_inches = "tight")
#plt.savefig("figures/figure2.png", bbox_inches = "tight")
plt.savefig("figures/figure3.png", bbox_inches = "tight")

plt.show()
```
